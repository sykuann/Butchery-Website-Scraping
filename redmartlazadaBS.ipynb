```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
```

```python
url = 'https://redmart.lazada.sg/shop-groceries-meat-seafood-fresh-meat-fresh-beef/?from=rm_nav_cate&m=redmart&page=3&spm=a2o42.lazmart_categories.lazmart_facet.1.49aa47b116JPOZ'
```

```python
driver = webdriver.Chrome()
```

```python
headers = ['product', 'price','weight', 'ori_price']
df = pd.DataFrame(columns = headers)
df.set_index('product', inplace = True)
```

```python
def findnumpage(soupie):
    toolbar = soupie.find_all('div', class_ = 'ant-col ant-col-20 ant-col-push-4 Jv5R8')
    for tb in toolbar:
        itemnumber = tb.find_all('span')
        pagenum = int(itemnumber[0].text.strip().split()[0])/30
        return(round(pagenum))
        
```

```python
def scrolldown():
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    match=False
    while(match==False):
            lastCount = lenOfPage
            time.sleep(2)
            lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
            if lastCount==lenOfPage:
                match=True
```

```python
driver.get(url)
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
pagenumber = findnumpage(soup)
```

```python
urlp = f'https://redmart.lazada.sg/shop-groceries-meat-seafood-fresh-meat-fresh-beef/?from=rm_nav_cate&m=redmart&page=8&spm=a2o42.lazmart_categories.lazmart_facet.1.49aa47b116JPOZ'
```

```python
driver.get(urlp)
time.sleep(4)
scrolldown()
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
a = soup.find_all('a', class_ = 'RedmartProductCard-link')
for h4 in a:
    h = h4.find('h4', class_ = 'RedmartProductCard-title').text.strip()
    weight = h4.find('div', class_ = 'RedmartProductCard-weight').text.strip()
    price = h4.find('div', class_ = 'RedmartProductCard-price sg').text.strip()
    df.loc[h,'weight']= weight 
    df.loc[h,'price']= price
    try:
        ori_price = h4.find('div', class_ = 'RedmartProductCard-originalPrice').text.strip()
        df.loc[h, 'ori_price']= ori_price
    except:
        continue
        
        
        
```
